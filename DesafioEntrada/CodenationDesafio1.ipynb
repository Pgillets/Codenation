```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.ensemble import RandomForestRegressor

df_train = pd.read_csv('train.csv').fillna(0)
df_test = pd.read_csv('test.csv').fillna(0)

df_test.columns.values
```

```python
columns_df_test = ['NU_INSCRICAO', 'CO_UF_RESIDENCIA', 'SG_UF_RESIDENCIA', 'NU_IDADE',
       'TP_SEXO', 'TP_COR_RACA', 'TP_NACIONALIDADE', 'TP_ST_CONCLUSAO',
       'TP_ANO_CONCLUIU', 'TP_ESCOLA', 'TP_ENSINO', 'IN_TREINEIRO',
       'TP_DEPENDENCIA_ADM_ESC', 'IN_BAIXA_VISAO', 'IN_CEGUEIRA',
       'IN_SURDEZ', 'IN_DISLEXIA', 'IN_DISCALCULIA', 'IN_SABATISTA',
       'IN_GESTANTE', 'IN_IDOSO', 'TP_PRESENCA_CN', 'TP_PRESENCA_CH',
       'TP_PRESENCA_LC', 'CO_PROVA_CN', 'CO_PROVA_CH', 'CO_PROVA_LC',
       'CO_PROVA_MT', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'TP_LINGUA', 'TP_STATUS_REDACAO', 'NU_NOTA_COMP1', 'NU_NOTA_COMP2',
       'NU_NOTA_COMP3', 'NU_NOTA_COMP4', 'NU_NOTA_COMP5',
       'NU_NOTA_REDACAO', 'Q001', 'Q002', 'Q006', 'Q024', 'Q025', 'Q026',
       'Q027', 'Q047']

df_train[columns_df_test]
```

```python
train_columns = ['CO_UF_RESIDENCIA', 'NU_IDADE',
       'TP_SEXO', 'TP_COR_RACA', 'TP_NACIONALIDADE', 'TP_ST_CONCLUSAO',
       'TP_ANO_CONCLUIU', 'TP_ESCOLA', 'TP_ENSINO', 'IN_TREINEIRO',
       'TP_DEPENDENCIA_ADM_ESC', 'IN_BAIXA_VISAO', 'IN_CEGUEIRA',
       'IN_SURDEZ', 'IN_DISLEXIA', 'IN_DISCALCULIA', 'IN_SABATISTA',
       'IN_GESTANTE', 'IN_IDOSO', 'TP_PRESENCA_CN', 'TP_PRESENCA_CH',
       'TP_PRESENCA_LC', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'TP_LINGUA', 'TP_STATUS_REDACAO', 'NU_NOTA_REDACAO']

x_treino = df_train[train_columns]
y_treino = df_train['NU_NOTA_MT']

x_treino
```

```python
x_treino['TP_SEXO'] = x_treino['TP_SEXO'].map({'M':1, 'F':0})

x_treino
```

```python
y_treino
```

```python
inscricao = df_test['NU_INSCRICAO']

df_test_clean = df_test[train_columns]
df_test_clean['TP_SEXO'] = df_test_clean['TP_SEXO'].map({'M':1, 'F':0})
df_test_clean
```

```python
random_forest = RandomForestRegressor(n_estimators=100)
random_forest.fit(x_treino, y_treino)
```

```python
y_pred = random_forest.predict(df_test_clean)
y_pred
```

```python
output = pd.DataFrame({'NU_INSCRICAO': inscricao, 'TP_PRESENCA_CN': df_test['TP_PRESENCA_CN'], 'NU_NOTA_MT': y_pred})
output
```

```python
# para cada ['TP_PRESENCA_CN'] == 0, tenho que transformar o ['NU_NOTA_MT'] em np.nan
i=3

for i in range(0,4576):
    if (output.iloc[i:1+i,1:2] == 0).bool(): #ver os ausentes
        output.iloc[i:1+i,2:3] = np.nan
        i+=3

output
```

```python
output = output.drop('TP_PRESENCA_CN', axis=1)
output
```

```python
output.to_csv('answer.csv', index=False)
```
